````python
import numpy as np
import pandas as pd

# Generate linear data
x = np.linspace(0, 10, 100)
y = 2.5 * x + np.random.normal(0, 2, size=x.shape)

# Create a DataFrame
linear_data = pd.DataFrame({'x': x, 'y': y})

# Save to CSV
linear_data.to_csv('linear_data.csv', index=False)

print("Linear dataset generated and saved as 'linear_data.csv'.")
````

````python
import pandas as pd
import numpy as np

# Categories
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = np.random.randint(50, 150, size=len(categories))

# Create a DataFrame
categorical_data = pd.DataFrame({'Category': categories, 'Value': values})

# Save to CSV
categorical_data.to_csv('categorical_data.csv', index=False)

print("Categorical dataset generated and saved as 'categorical_data.csv'.")
````

````python
import numpy as np
import pandas as pd

# Generate bivariate data
np.random.seed(42)
x = np.random.normal(50, 15, 300)
y = 0.5 * x + np.random.normal(0, 7, size=x.shape)

# Create a DataFrame
bivariate_data = pd.DataFrame({'x': x, 'y': y})

# Save to CSV
bivariate_data.to_csv('bivariate_data.csv', index=False)

print("Bivariate dataset generated and saved as 'bivariate_data.csv'.")
````

````python
### **d. Generating a Time Series Dataset for Line Plots**
import pandas as pd
import numpy as np

# Generate date range
dates = pd.date_range(start='2021-01-01', periods=365, freq='D')

# Generate time series data
np.random.seed(0)
values = np.random.normal(0, 1, size=len(dates)).cumsum() + 100

# Create a DataFrame
timeseries_data = pd.DataFrame({'Date': dates, 'Value': values})

# Save to CSV
timeseries_data.to_csv('timeseries_data.csv', index=False)

print("Time series dataset generated and saved as 'timeseries_data.csv'.")
````

````python
### **e. Generating a Multivariate Dataset for Heatmaps**
import pandas as pd
import numpy as np

# Generate random data
np.random.seed(1)
data = np.random.rand(100, 5)

# Create a DataFrame
columns = ['Feature1', 'Feature2', 'Feature3', 'Feature4', 'Feature5']
multivariate_data = pd.DataFrame(data, columns=columns)

# Save to CSV
multivariate_data.to_csv('multivariate_data.csv', index=False)

print("Multivariate dataset generated and saved as 'multivariate_data.csv'.")
````

````python



### **b. Visualizing the Categorical Dataset**
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
categorical_data = pd.read_csv('categorical_data.csv')

# Bar plot using Seaborn
plt.figure(figsize=(8, 6))
sns.barplot(x='Category', y='Value', data=categorical_data, palette='Set2')
plt.title('Categorical Data Visualization')
plt.xlabel('Category')
plt.ylabel('Value')
plt.savefig('categorical_plot.png', dpi=300)
plt.show()
```

### **c. Visualizing the Bivariate Dataset**
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
bivariate_data = pd.read_csv('bivariate_data.csv')

# Scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.scatterplot(x='x', y='y', data=bivariate_data, alpha=0.5)
sns.regplot(x='x', y='y', data=bivariate_data, scatter=False, color='r')
plt.title('Bivariate Data Visualization')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.savefig('bivariate_plot.png', dpi=300)
plt.show()
```

### **d. Visualizing the Time Series Dataset**
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
timeseries_data = pd.read_csv('timeseries_data.csv')

# Convert 'Date' to datetime
timeseries_data['Date'] = pd.to_datetime(timeseries_data['Date'])

# Line plot
plt.figure(figsize=(12, 6))
plt.plot(timeseries_data['Date'], timeseries_data['Value'], color='g')
plt.title('Time Series Data Visualization')
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('timeseries_plot.png', dpi=300)
plt.show()
```

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
multivariate_data = pd.read_csv('multivariate_data.csv')

# Compute correlation matrix
corr_matrix = multivariate_data.corr()

# Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.savefig('heatmap_plot.png', dpi=300)
plt.show()
```


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Generate linear data
x = np.linspace(0, 10, 100)
y = 2.5 * x + np.random.normal(0, 2, size=x.shape)

# Create a DataFrame
linear_data = pd.DataFrame({'x': x, 'y': y})

# Line plot
plt.figure(figsize=(10, 6))
plt.plot(linear_data['x'], linear_data['y'], marker='o', linestyle='-', color='b')
plt.title('Linear Relationship')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid(True)
plt.show()
```


import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
bivariate_data = pd.read_csv('bivariate_data.csv')

# Scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.scatterplot(x='x', y='y', data=bivariate_data, alpha=0.5)
sns.regplot(x='x', y='y', data=bivariate_data, scatter=False, color='red')
plt.title('Relationship between X and Y')
plt.xlabel('Variable X')
plt.ylabel('Variable Y')
plt.savefig('bivariate_story_plot.png', dpi=300)
plt.show()



````
